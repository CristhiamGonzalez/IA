## Primer punto Distribuciones discretas.
1. Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números: 
1. Grafique los números aleatorios,

```python
#Para todas las simulaciones:
N=100000 # Tamaño de muestras
m=3000  # Anchos de secuencias
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# bernoulli con cierto p entre 0 y 1
p = 0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
print ("Media: ",mean, "Varianza: ",var, "F1: ",skew, "F2: ",kurt)
r = bernoulli.rvs(p, size=N) #Generate random numbers
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
```

```python
r.sort() #Ordenar
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
```

```python
plt.hist(r, bins = m) #Histograma, "r" son los números aleatorios de bernoulli y "m" son los anchos iguales de bins o secuencias 
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean() #Media de los números aleatorios
varest=r.var() #La varianza de los números aleatorios
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```python
# binomial con cierto p entre 0 y 1 y cierto n
n,p = 10, 0.5
from scipy.stats import binom
meanb, varb, skewb, kurtb = binom.stats(n, p, moments='mvsk')
print ("Media: ",meanb, "Varianza: ",varb , "F1: ",skewb, "F2: ",kurtb)
rb = binom.rvs(n, p, size=N) #Generate random numbers
plt.plot(rb)
plt.title(str(N)+" puntos distribuidos binomiales con p="+str(p)+" y con n="+str(n))
plt.show()
```

```python
rb.sort() #Ordenar
plt.plot(rb)
plt.title(str(N)+" puntos distribuidos binomiales ordenados con p="+str(p)+" y con n="+str(n))
plt.show()
```

```python
plt.hist(rb, bins = m) #Histograma, "rb" son los números aleatorios binomial y "m" son los anchos iguales de bins o secuencias 
plt.title("Histograma de los "+str(N)+" puntos distribuidos binomiales con p="+str(p)+" y con n="+str(n))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo rb:")
print(type(rb))
print("Vamos a hallar algunas estimaciones de estadisticas de rb")
avgb=rb.mean() #Media de los números aleatorios
varestb=rb.var() #La varianza de los números aleatorios
print("Los valores teoricos de la media y varianza para la binomiales con p=",p,"son ub=" ,meanb,"varb=",varb)
print("Los valores estimados en nuestro experimento de la media y varianza para la binomiales con p=",p,"son u_estb=" ,avgb,"var_estb=",varestb)
```

## Segundo punto Distribuciones contínuas.


```python
#Para todas las simulaciones:
N=1000
m=300
numpy.random.seed(seed=1314926535)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# Normal de media 5 y desviación estandar 0.5
u=5
std=0.5
mean, var, skew, kurt = scipy.stats.norm.stats(moments='mvsk',loc=u,scale=std)
r = scipy.stats.norm.rvs (loc=u,scale=std, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales con media="+str(u)+" y std="+str(std))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos normales ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normales con u="+str(u)+ " y std= "+str(std))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
# Uniforme entre 0 y 255
ini = 0
fin = 255
from scipy.stats import uniform
mean, var, skew, kurt = uniform.stats(moments='mvsk',loc=ini,scale=fin)
r = scipy.stats.uniform.rvs (loc=ini,scale=fin, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos uniformes entre "+str(ini)+" y "+str(fin))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos uniformes ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos uniformes entre "+str(ini)+ " y "+str(fin))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la uniforme son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
# Exponencial con Lamnda L
L=0.3
ini = 0
fin = 1/L
from scipy.stats import expon
mean, var, skew, kurt = expon.stats(moments='mvsk',loc=ini,scale=fin)
r = scipy.stats.expon.rvs (loc=ini,scale=fin, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos exponenciales con Lamnda "+str(L))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos exponenciales ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos exponenciales con Lamnda "+str(L))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la exponencial son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
#Chi cuadrada con K grados de libertad
K=3.0
from scipy.stats import chi2
mean, var, skew, kurt = chi2.stats(K, moments='mvsk')
r = scipy.stats.chi2.rvs (K, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos Chi cuadrado con "+str(K)+ "grados de libertad")
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos chi cuadrado)
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos Chi cuadrado con "+str(K)+ " grados de libertad")
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la chi cuadrado son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
#Gamma con parametros k y theta t
k,t=2.54,0.43
ini = 0
fin = 1/t
from scipy.stats import gamma
mean, var, skew, kurt = gamma.stats(k, loc = ini, scale =fin, moments='mvsk')
r = scipy.stats.gamma.rvs (k, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos gamma con k "+str(k)+ " y theta de "+str(t))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos gamma)
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos gamma con k "+str(k)+ " y theta de "+str(t))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para gamma son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```
